```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("sonar_data.csv",header=None)
df.shape
```

```python
df.head(10)
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

```python
# separating data and labels
X = df.drop(columns=60,axis =1)
Y = df[60]
```

```python
# split into training and test data

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(x_train.shape, x_train.shape, x_test.shape)
```

```python
#Model -> Logistic Regression

model= LogisticRegression()

model.fit(x_train,y_train)
```

```python
#model evaluation or accuracy of the training data

x_train_prediciton=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediciton,y_train)
```

```python
print(training_data_accuracy)
```

```python
#model evaluation or accuracy of the test data

x_test_prediciton=model.predict(x_test)
training_data_accuracy=accuracy_score(x_test_prediciton,y_test)

print(training_data_accuracy)
```

```python
#making a prediction system
input_data= X.iloc[4]

#changing input data to np array
input_data_as_np_array=np.asarray(input_data)

#reshape the np array as we are predicting for one instance
input_data_reshaped = input_data_as_np_array.reshape(1,-1)

prediciton = model.predict(input_data_reshaped)

print(prediciton)
```
